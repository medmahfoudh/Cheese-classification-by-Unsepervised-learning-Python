**L'importation du Bibliothéque Numpy:C'est une Bibliothéque dispose d'un grand nombre de fonctions mathématiques qui peuvent être appliquées directement à un tableau.**

**et on fait le clustering a l'aide de l'algorithme de k-means, on utilise le module "cluster" dans le bibliothéque "Sickit-learn" ou sklearn.**

```python
import numpy as np 
from sklearn import cluster
np.random.seed(0) #si on utilise random.seed() On évite le changement des aléatoires à chaque fois, dans notre cas pour éviter le changement des centres de clusters à chaque fois qu'ona exécuter le code
kmeans = cluster.KMeans(n_clusters = 4) #on utilise le fonction cluster.KMeans() Pour donne les nombres de k(Les Centres de clusters)
kmeans.fit(fromage) #Pour training notre modéle on utilise kmeans.fit()
#Aprés l'utilisation de fit() notre modèle est prêt
#Et nous obtenons des étiquettes pour ces clusters en utilisant kmeans.labels_
idk = np.argsort(kmeans.labels_) #Pour Trier les étiquette selon Les Classes
print(pd.DataFrame(fromage.index[idk],kmeans.labels_[idk])) #Afficher les Fromages aprés le clustering et ses clusters...
print(kmeans.cluster_centers_) #Afficher les centres de clusters
```

**L'importation du module metrics: Le module sklearn.metrics comprend des fonctions de score, des mesures de performance et des métriques par paires et des calculs de distance.**

```python
from sklearn import metrics
#utilisation de la métrique "silhouette" 
res = np.arange(9, dtype ="double") #Renvoie un tableau avec des éléments régulière, cette fonction prend Le Debut e l'arrêt et le datatypes 
for k in np.arange(9): #pour faire un boucle et a chaque index on le met l'accuracy dans 'res' (ou le table des erreurs de chaque cluster)
    km = cluster.KMeans(n_clusters = k+2)
    km.fit(fromage)
    res[k] = metrics.silhouette_score(fromage,km.labels_)
print (res)
#l'affichage du graphe
import matplotlib.pyplot as plt
plt.title("silhouette")
plt.xlabel("# of clusters")
plt.plot(np.arange(2,11,1),(res))
plt.show
```

**- L'importation du matplotlib: C'est un Bibliothéque Python permet de visualiser les données**
**- L'importation du dendogram , linkage et fcluster à partir de scipy.cluster.hierarchy:
   - linkage: Effectuer un clustering hiérarchique.
   - fcluster: Formez des grappes plates à partir du clustering hiérarchique défini par la matrice de liaison donnée.
   - dendogram: Pour Tracez le clustering hiérarchique sous forme de dendrogramme.

```python
#librairies pour la CAH
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
Z = linkage(fromage,method='ward', metric='euclidean') # Pour faire le clustering hierarchical, cette fonction prend le datasets, le methode=ward et le metric=euclidiean c'est à dire on faire le clustering hierarchical en utilisant la distance euclidienne
#affichage du dendrogramme
plt.title('CAH avec matérialisation des 4 classes')  #Le Titre du shéma
plt.figure(figsize = (16,8))
dendrogram(Z,labels=fromage.index ,color_threshold=255) #On utilise le fonction dendogram pour tracez le dendograme, et on les étiquette et l'orientation de notre figure
plt.show() #Fonction dans matplotlib permet d'afficher le figure  
groupes_cah = fcluster(Z, t = 255,criterion='distance') #Les groupes de clustering
print(groupes_cah) #L'affichage des groupes de clustering 
#index triés des groupes
idg = np.argsort(groupes_cah) #Trier les index de groupes clustering 
#affichage des observations et leurs groupes
print(pd.DataFrame(fromage.index[idg],groupes_cah[idg]))
```

```python
#correspondance avec les groupes de la CAH
pd.crosstab(groupes_cah,kmeans.labels_) #Cette méthode est utilisée pour calculer une simple tabulation croisée de deux facteurs (ou plus), dans notre cas on a le groupe hierarchical et les clusters de k.means
```

**Avec l'PCA, nous tenon compte des liaisons entre les variables**

```python
from sklearn.decomposition import PCA #sklearn.decomposition.PCA Permet de faire l’analyse en composantes principales 
acp = PCA(n_components =2).fit_transform(fromage)
#Pour Afficher le graphe 
for couleur,k in zip(['red','blue','lawngreen', 'aqua'],[0,1,2,3]):
     plt.scatter(acp[kmeans.labels_==k,0],acp[kmeans.labels_==k,1],c=couleur)
plt.show()
```

```python
from sklearn.decomposition import PCA #sklearn.decomposition.PCA Permet de faire l’analyse en composantes principales 
acp = PCA(n_components =2).fit_transform(fromage)
#Pour Afficher graphe 
for couleur,k in zip(['red','blue','lawngreen', 'aqua'],[1,2,3,4]):
     plt.scatter(acp[groupes_cah==k,0],acp[groupes_cah==k,1],c=couleur)
plt.show()
```

### selon l'analyse on trouve que CAH donne le meilleur performance

# 2) L'utilisation du sklearn.cluster.AgglomerativeClustering


**L'importation du module AgglomerativeClustering: c'est un type de clustering hiérarchique qui fonctionne de manière ascendante.**

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import fcluster
import numpy as np
#import sklearn.metrics as sm
cah = AgglomerativeClustering(n_clusters=4, linkage='ward').fit(fromage)

print(cah.labels_)
idx = np.argsort(cah.labels_) #Trier les index de groupes clustering 
#affichage des observations et leurs groupes
print(pd.DataFrame(fromage.index[idx],cah.labels_[idx]))
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
linkage_matrix = linkage(fromage , 'ward')
plot = plt.figure(figsize = (14 , 7))
dendrogram(linkage_matrix,labels=fromage.index , color_threshold=255)
plt.title("Test")
plt.show()
```

```python
pd.crosstab(cah.labels_,kmeans.labels_, rownames=['cah'] , colnames=['kmeans'])
```

```python
#=========AgglomerativeClustering========

from sklearn.decomposition import PCA #sklearn.decomposition.PCA Permet de faire l’analyse en composantes principales 
acp = PCA(n_components =2).fit_transform(fromage)
#Pour Afficher le graphe 
for couleur,k in zip(['red','blue','lawngreen', 'aqua'],[0,1,2,3]):
     plt.scatter(acp[cah.labels_==k,0],acp[cah.labels_==k,1],c=couleur)
plt.title("AgglomerativeClustering")
plt.show()

#==========Kmenas==========#
for couleur,k in zip(['red','blue','lawngreen', 'aqua'],[0,1,2,3]):
     plt.scatter(acp[kmeans.labels_==k,0],acp[kmeans.labels_==k,1],c=couleur)
plt.title("Kmeans")
plt.show()

#=========CAH Scipy==========#
for couleur,k in zip(['red','blue','lawngreen', 'aqua'],[1,2,3,4]):
     plt.scatter(acp[groupes_cah==k,0],acp[groupes_cah==k,1],c=couleur)
plt.title("CAH Scipy")
plt.show()
```
